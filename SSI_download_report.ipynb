```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
#import re
```

```python

req = Request("https://www.ssi.dk/sygdomme-beredskab-og-forskning/sygdomsovervaagning/c/covid19-overvaagning")
html_page = urlopen(req)

soup = BeautifulSoup(html_page, "lxml")
```

```python
soup
```

```python
links = []
for link in soup.findAll('a'):
    links.append(link.get('href'))

print(links)
```

```python
for link in links:
    if isinstance(link, str) and "Data-Epidemiologiske-Rapport" in link:
        print(link)
        
#C:\Users\boj\Downloads\Data-Epidemiologiske-Rapport-29102020-mj23
```
